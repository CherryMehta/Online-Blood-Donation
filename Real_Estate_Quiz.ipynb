```
gurgaon["CITY"] = "Gurgaon"
hyderabad["CITY"] = "Hyderabad"
kolkata["CITY"] = "Kolkata"
mumbai["CITY"] = "Mumbai"
```

```
df = pd.concat([gurgaon, hyderabad, kolkata, mumbai], ignore_index=True)
```

```
useful_cols = [
    "CITY",
    "PROPERTY_TYPE",
    "LOCALITY",
    "PRICE",
    "BEDROOM_NUM",
    "AREA",
    "PRICE_PER_UNIT_AREA",
    "FURNISH",
    "TRANSACT_TYPE"
]
```

```
df = df[[col for col in useful_cols if col in df.columns]]
df = df.dropna()
```

```
final_df = pd.concat(
    [gurgaon, hyderabad, kolkata, mumbai],
    ignore_index=True
)
```

```
final_df.shape
final_df["CITY"].value_counts()
final_df.head()
```

```
def row_to_text(row):
    return (
        f"City: {row['CITY']}, "
        f"Locality: {row['LOCALITY']}, "
        f"Property Type: {row['PROPERTY_TYPE']}, "
        f"Bedrooms: {row['BEDROOM_NUM']}, "
        f"Area: {row['AREA']}, "
        f"Price: {row['PRICE']}, "
        f"Transaction Type: {row['TRANSACT_TYPE']}, "
        f"Furnishing: {row['FURNISH']}"
    )

final_df["text_data"] = final_df.apply(row_to_text, axis=1)
```

```
knowledge_base = "\n".join(final_df["text_data"].sample(500))
```
